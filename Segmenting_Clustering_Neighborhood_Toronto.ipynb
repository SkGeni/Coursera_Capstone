```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json 

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt

# import k-means from clustering stage
from sklearn.cluster import KMeans

import folium # map rendering library

#for web-scrapping
import requests
import lxml.html as lh

print('Libraries imported.')
```

Hitting the url and getting all the rows present on that page

```python
#url of wikipedia page
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url)
doc = lh.fromstring(page.content)
tr_elements = doc.xpath('//tr')
```

Creating a 2d list from the table scrapped by the Wikipedia page.

```python
#initializing the list
data=[]

#for each row, we are iterating to each element and appending it to the 2d list data
for t in tr_elements[1:]:
    name=[]
    for T in t:
        element=T.text_content().rstrip()
        name.append(element)
    data.append(name)
```

Creating the dataframe *init_df* and cleaning

```python
#defining the columns of the data frame
col=['PostalCode', 'Borough', 'Neighborhood']

#creating the pandas dataframe from data tables
init_df = pd.DataFrame(data[:-4], columns=col)

#dropping the rows having the entry "Not assigned" in Borough
init_df = init_df[~init_df.Borough.str.contains('Not assigned')]

#resetting index of the dataframe
init_df.reset_index(drop=True, inplace=True)

#dropping duplicates
init_df.drop_duplicates()

#where neighborhood is not assigned or null the value of Borough is inserted for that row
init_df['Neighborhood'].replace({'Not assigned':init_df['Borough'], "":init_df['Borough']}, inplace=True)
```

<h3>
Part 2: Latitude and Longitude columns in Dataframe
</h3>

For each neighborhood we need to find latitude and longitude, we have a csv file provided all the latitudes and longitudes of Postal codes of Neighborhood. So we are reading from the csv file and renaming the column 'Postal Code' to 'PostalCode' as it is the column name present in the init_df

```python
#reading the csv as pandas dataframe
!wget -q -O 'Geospatial_Coordinates.csv' http://cocl.us/Geospatial_data
coordinates = pd.read_csv('Geospatial_Coordinates.csv')
#renaming the column as needed
coordinates.rename(columns={'Postal Code':'PostalCode'},inplace=True)
```

Merging the dataframes based on _PostalCode_ using outer join.

```python
df_cor = pd.merge(init_df, coordinates, how='outer', on=['PostalCode'])
```

<h3>
    Part 3 : Exploring and Clustering the neighborhoods in Toronto.
</h3>

I am only taking the colomn 'Borough' having *Toronto*

```python
df_relevant = df_cor[df_cor['Borough'].str.contains("Toronto")]
```

Dropping the indices and getting the *df_relevant* dataset

```python
df_relevant.reset_index(drop=True, inplace=True)
```

Just to quickly check the relevant dataset having Borough as Toronto

```python
df_relevant.head()
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, neighborhood, borough in zip(df_relevant['Latitude'], df_relevant['Longitude'], df_relevant['Neighborhood'], df_relevant['Borough']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_toronto)
map_toronto
```
